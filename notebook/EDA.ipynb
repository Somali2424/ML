```python
math_average=df.groupby(by='gender')['math_score'].mean()
print(math_average)
```

```python
total_average=df.groupby(by='gender')['total score'].mean()
print(total_average)
```

```python
n_groups = 2
men_averages = (total_average[1],math_average[1])
women_averages = (total_average[0],math_average[0])

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, men_averages, bar_width,
alpha=opacity,
color='g',
label='Men')

rects2 = plt.bar(index + bar_width, women_averages, bar_width,
alpha=opacity,
color='r',
label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.xticks(index + bar_width, ("total average", "math average"))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#On an average females have a better overall score than men.
#whereas males have scored higher in Math
```
